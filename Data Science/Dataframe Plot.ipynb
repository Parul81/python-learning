# DataFrame Plot :-

```python
import pandas as pd
```

```python
data = {"Name":["Sean","Jack","Liam","Alex"],
       "Math":[96,85,75,100]}
```

```python
data
```

```python
type(data)
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
## How to Make Pie Chart
```

```python
df.plot.pie(title="Stud_Marks",y="Math")
```

```python
## Options of The Pie Chart
```

```python
## figsize() :

    ## Syntax :  figsize((width,height))
```

```python
df.plot.pie(title="Stud_Marks",y="Math",figsize=(5,5))
```

```python
## fontsize :

    ## Set For Labels
```

```python
df.plot.pie(title="Stud_Marks",y="Math",fontsize=15)
```

```python
## labels :

  ## Values ==> List()
    
    ## Syntax : df_name.plot.pie(title,y,figsize,fonsize,labels=[values])
```

```python
labl = ["Sean","Jack","Liam","Alex"]
df.plot.pie(title="Stud_Marks",y="Math",fontsize=15,labels=labl)
```

```python
### labeldistance :

  ## Bydefault value = 1.1
```

```python
df.plot.pie(title="Stud_Marks",y="Math",fontsize=15,labels=labl,labeldistance=2)
```

```python
df.plot.pie(title="Stud_Marks",y="Math",fontsize=15,labels=labl,labeldistance=1.5)
```

```python
### legend :

  ## Bydefault => True
```

```python
df.plot.pie(title="Stud_Marks",y="Math",fontsize=15,labels=labl,labeldistance=2,legend=False)
```

```python
## explode()
```

```python
exp = (0.1,0,0,0)
df.plot.pie(title="Stud_Marks",y="Math",explode=exp)
```

```python
exp = (0.1,0,0.1,0)
df.plot.pie(title="Stud_Marks",y="Math",explode=exp)
```

```python
exp = (1.2,0,0,0)
df.plot.pie(title="Stud_Marks",y="Math",explode=exp)
```

```python
## Startangle
```

```python
exp = (0,0,0,0)
df.plot.pie(title="Stud_Marks",y="Math",explode=exp,startangle=40)
```

```python
df.plot.pie(title="Stud_Marks",y="Math",explode=exp,startangle=100)
```

```python
df.plot.pie(title="Stud_Marks",y="Math",explode=exp,startangle=270)
```

```python
df.plot.pie(title="Stud_Marks",y="Math",explode=exp,startangle=-40)
```

```python
df.plot.pie(title="Stud_Mark",y="Math",colors=["g","cyan","orange","r"],startangle=90)
```

```python
### counterclock :

  ## Bydefault Value : True
    
    ##  Direction os Fractions
```

```python
df.plot.pie(title="Stud_Mark",y="Math",colors=["g","cyan","orange","r"],startangle=90,counterclock=False)
```

```python
## rotateklabels:
```

```python
df.plot.pie(title="Stud_Mark",y="Math",rotatelabels=[45])
```

```python
df.plot.pie(title="Stud_Marks",y="Math",labels=labl,rotatelabels=[45])
```

```python
df.plot.pie(title="Stud_Marks",y="Math",labels=labl,rotatelabels=[45,90])
```

```python
df.plot.pie(title="Stud_Marks",y="Math",labels=labl,rotatelabels=(45,90,135,45))
```

```python
df.plot.pie(title="Stud_Marks",y="Math",labels=labl,rotatelabels=True)
```

```python
## shadow:
```

```python
df.plot.pie(title="Stud_Marks",y="Math",labels=labl,shadow=True)
```

```python
## autopct :

 ## It Is Used To Display The Percentage of Each Segment ,
    
 ## We Placed The One Decimal 

 ## autopct = "%1.1f%%"
    
 ## Placed The Two Decimal

 ## autopct = "%1.2f%%"
```

```python
df.plot.pie(title="Stud_Marks",y="Math",autopct='%1.1f%%')
```

```python
df.plot.pie(title="Stud_Marks",y="Math",autopct='%1.2f%%')
```

```python
## frame :
  
             ## ByDefault => False
```

```python
df.plot.pie(title="Stud_Marks",y="Math",autopct='%1.1f%%',frame=True)
```

```python
## How To Save The Graph In Image File
  
    ## Object_name.get_figure()
    
    ## fig = Object_name.get_figure()
    
    ## fig.savefig("File_Name")
```

```python
plot = df.plot.pie(title="Stud_Marks",y="Math",autopct='%1.1f%%')
fig = plot.get_figure()
fig.savefig("Ks1.jpg")
```

```python
from PIL import Image

img = Image.open("ks1.jpg")
im1 = Image.open('ks.jpg')

im1.save("out.gif", save_all=True, append_images=[img, im1], duration=100,loop=0)
```
