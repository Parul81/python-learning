```python
import numpy as np
```

```python
x = np.arange(0,np.pi*2,0.05)
y = np.sin(x)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
## Set The Title axes
ax.plot(x,y)
ax.set_title("Sine Wave")
ax.set_xlabel("Angle")
ax.set_ylabel("Sine")
plt.show()
```

### How To Make Image of Graph / Save Graph As Image
### Use : savefig
### Syntax : plt.savefig("file_name.extension")

```python
x = np.arange(0,np.pi*2,0.05)
y = np.sin(x)
y1 = np.cos(x)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
## Set The Title axes
ax.plot(x,y)
ax.plot(x,y1)
ax.plot(x,np.sin(-x))
ax.set_title("Sine Wave")
ax.set_xlabel("Angle")
ax.set_ylabel("Sine")
## Save Graph As An Image File
plt.savefig("Graph.png")
plt.show()
```

```python
def g1():
    x = np.arange(0,np.pi*2,0.05)
    y = np.sin(x)
    y1 = np.cos(x)
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    ## Set The Title axes
    ax.plot(x,y)
    ax.plot(x,y1)
    ax.plot(x,np.sin(-x))
    ax.plot(x,np.tan(x))
    ax.set_title("Sine Wave")
    ax.set_xlabel("Angle")
    ax.set_ylabel("Sine")
    plt.show()
```
