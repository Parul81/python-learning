```python
import numpy as np
x = np.arange(0,np.pi*2,0.05)
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) ## Main Axes
axes2 = fig.add_axes([0.1,0.1,0.3,0.3])  ## Inset Axis
y = np.sin(x)
axes1.plot(x,y,"g")
axes2.plot(x,np.cos(x),"orange")
axes1.set_title("Sine")
axes2.set_title("Cosine")
plt.show()
```
