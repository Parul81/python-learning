```python
## xlabel
## ylabel
```

```python
x = np.arange(0,np.pi*2,0.05)
y = np.sin(x)
plt.xlabel("Angle")
plt.ylabel("Sine")
plt.title("Sine Wave")
plt.plot(y)
plt.show
```

```python
x = np.arange(0,np.pi*2,0.05)
y = np.sin(x)
plt.xlabel("Angle")
plt.ylabel("Sine")
plt.title("Sine Wave")
plt.plot(x,y,color="pink")
plt.show
```

```python
x = np.arange(0,np.pi*2,0.05)
y = np.sin(x)
plt.xlabel("Angle")
plt.ylabel("Sine")
plt.title("Sine Wave")
plt.plot(x,-y,color="red")
plt.show
```
