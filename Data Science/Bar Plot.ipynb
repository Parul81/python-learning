```python
rate1 = [20,29,35,36,45]
rate2 = [25,20,30,45,44]
x = np.arange(5)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(x,rate1,width =0.2,color="g")
ax.bar(x,rate2,width=0.2,color="orange")
ax.set_ylabel("Price")
ax.set_title("Product")
ax.set_xticks(x,["DAY1","DAY2","DAY3","DAY4","DAY5"])
ax.set_yticks(np.arange(0,50,10))
ax.legend(labels=["Diesel","Petrol"])
plt.show()
```

```python
rate1 = [20,29,35,36,45]
rate2 = [25,20,30,45,44]
x = np.arange(5)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(x,rate1,width =0.2,color="g")
ax.bar(x,rate2,width=0.2,bottom=rate1,color="orange")
ax.set_ylabel("Price")
ax.set_title("Product")
ax.set_xticks(x,["DAY1","DAY2","DAY3","DAY4","DAY5"])
ax.set_yticks(np.arange(0,50,10))
ax.legend(labels=["Diesel","Petrol"])
plt.show()
```
