```python
import matplotlib.pyplot as plt

y = [1,4,9,16,25,36,49,64]

x1 = [1,16,30,40,55,66,100,105]

x2 = [1,5,6,10,12,15,55,75]

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

l1 = ax.plot(x1,y,'gs--')

l2 = ax.plot(x2,y,'co-')

ax.legend(labels = ("HeadPhone","SmartPhone"))

ax.set_title("Sales Effect")

ax.set_xlabel("Gadgets")

ax.set_ylabel("Sales")

plt.show()

##loc='lower right'
```

```python
import matplotlib.pyplot as plt

y = [1,4,9,16,25,36,49,64]

x1 = [1,16,30,40,55,66,100,105]

x2 = [1,5,6,10,12,15,55,75]

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

l1 = ax.plot(x1,y,'gs--')

l2 = ax.plot(x2,y,'co-')

ax.legend(labels = ("HeadPhone","SmartPhone"),loc='center')

ax.set_title("Sales Effect")

ax.set_xlabel("Gadgets")

ax.set_ylabel("Sales")

plt.show()
```

```python
import matplotlib.pyplot as plt
y = [1,4,9,16,25,36,49,64]
x1 = [1,16,30,40,55,66,100,105]
x2 = [1,5,6,10,12,15,55,75]
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
l1 = ax.plot(x1,y,'gs--')
l2 = ax.plot(x2,y,'co-')
ax.legend(labels = ("HeadPhone","SmartPhone"),loc=5)
ax.set_title("Sales Effect")
ax.set_xlabel("Gadgets")
ax.set_ylabel("Sales")
plt.show()
```
