# Twin_Axes :-

```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
a1 = fig.add_axes([0,0,1,1])
x = np.arange(1,11)
a1.plot(x,np.exp(x))
a1.set_ylabel("Exp")
a2 = a1.twinx()
a2.plot(x,np.log(x),"r--")
a2.set_ylabel("Log")
fig.legend(labels=("Exp","Log"))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
a1 = fig.add_axes([0,0,1,1])
x = np.arange(1,11)
a1.plot(x,np.exp(x))
a1.set_ylabel("Exp")
a2 = a1.twinx()
a2.plot(x,np.log(x),"r--")
a5 = a2.twinx()
a5.plot(x,np.sin(x),'g-.')
a2.set_ylabel("Log")
fig.legend(labels=("Exp","Log"),loc="upper left")
plt.show()
```
