# Scatter_Plot :-

```python
import matplotlib.pyplot as plt
virat = [100,95,120,56,85]
kane =  [120,98,95,85,156]
range_ = [25,50,75,100,200]
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.scatter(range_,virat,color="g")
ax.scatter(range_,kane,color="orange")
ax.set_xlabel("Runs")
ax.set_ylabel("ScoreBoard")
ax.set_title("Virat vs Kane")
plt.show()
```
