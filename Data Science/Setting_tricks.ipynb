```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0,np.pi*2,0.05)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
y = np.sin(x)
ax.plot(x,y)
ax.set_xlabel("Angle")
ax.set_title("Sine")
ax.set_xticks([0,2,4,6])
ax.set_xticklabels(["Zero","Two","Four","Six"])
ax.set_yticks([-1,0,1])
plt.show()
```
