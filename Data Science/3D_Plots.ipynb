```python
### SurFace Graph
```

```python
x = np.outer(np.linspace(-2,2,10),[1,2,3,4,5,6,7,8,10])
y=x.copy()
z=np.cos(x**2+y**2)
fig = plt.figure()
ax = plt.axes(projection="3d")
ax.plot_surface(x,y,z)
ax.set_title("Surface Plot")
plt.show()
```

```python
x = np.outer([1,2,3,4,5,6,7,8,10],np.linspace(-2,2,10))
y=x.copy()
z=np.cos(x**2+y**2)
fig = plt.figure()
ax = plt.axes(projection="3d")
ax.plot_surface(x,y,z)
ax.set_title("Surface Plot")
plt.show()
```

```python
x = np.outer(np.linspace(-2,2,10),np.linspace(-2,2,10))
y=x.copy().T
z=np.cos(x**2+y**2)
fig = plt.figure()
ax = plt.axes(projection="3d")
ax.plot_surface(x,y,z)
ax.set_title("Surface Plot")
plt.show()
```

```python
x = np.outer(np.linspace(-2,2,10),np.linspace(-2,2,10))
y=x.copy().T
z=np.cos(x**2+y**2)
fig = plt.figure()
ax = plt.axes(projection="3d")
ax.plot_surface(x,y,z,edgecolor="green")
ax.set_title("Surface Plot")
plt.show()
```

```python
x = np.outer(np.linspace(-2,2,10),np.linspace(-2,2,10))
y=x.copy().T
z=np.cos(x**2+y**2)
fig = plt.figure()
ax = plt.axes(projection="3d")
ax.plot_surface(x,y,z,cmap="viridis",edgecolor="green")
ax.set_title("Surface Plot")
plt.show()
```

```python
x = np.outer(np.linspace(-3, 3, 32), np.ones(32))
y = x.copy().T 
z = (np.sin(x **2) + np.cos(y **2) )
fig = plt.figure(figsize =(14, 9))
ax = plt.axes(projection ='3d')
my_cmap = plt.get_cmap('hot')
surf = ax.plot_surface(x, y, z,cmap = my_cmap,edgecolor ='green')
fig.colorbar(surf, ax = ax,shrink = 0.5, aspect = 5)
ax.set_title('Surface plot')
```
