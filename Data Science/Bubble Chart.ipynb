# Bubble Chart :-

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
plt.scatter(x,y,z)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
colors = np.random.rand(50)
plt.scatter(x,y,s=z*500,c=colors)
plt.show()
```
