```python
import tkinter as tk
from tkinter import messagebox

def open_rate_page():
    # Create a new window for the Rate Us page
    rate_window = tk.Toplevel(root)
    rate_window.title("Rate Us")

    # Label and entry for rating
    rate_label = tk.Label(rate_window, text="Please rate us (1 to 5):")
    rate_label.pack(pady=5)
    
    rate_var = tk.StringVar()
    rate_entry = tk.Entry(rate_window, textvariable=rate_var)
    rate_entry.pack(pady=5)

    # Label and text box for comments
    comments_label = tk.Label(rate_window, text="Comments:")
    comments_label.pack(pady=5)
    
    comments_text = tk.Text(rate_window, height=5, width=30)
    comments_text.pack(pady=5)

    # Function to handle the submission
    def submit_rating():
        rating = rate_var.get()
        comments = comments_text.get("1.0", tk.END).strip()

        if rating.isdigit() and 1 <= int(rating) <= 5:
            messagebox.showinfo("Thank You", "Thank you for your feedback!")
            rate_window.destroy()
        else:
            messagebox.showerror("Invalid Input", "Please enter a valid rating between 1 and 5.")

    # Submit button
    submit_button = tk.Button(rate_window, text="Submit", command=submit_rating)
    submit_button.pack(pady=10)

# Create the main window
root = tk.Tk()
root.title("Main Application")

# Create a menu bar
menu_bar = tk.Menu(root)

# Add the "Rate Us" option to the menu
rate_menu = tk.Menu(menu_bar, tearoff=0)
rate_menu.add_command(label="Rate Us", command=open_rate_page)
menu_bar.add_cascade(label="Help", menu=rate_menu)

# Attach the menu bar to the main window
root.config(menu=menu_bar)

# Run the application
root.mainloop()
```
