```python
import matplotlib.pyplot as plt
import numpy as np
fig,axes = plt.subplots(1,2,figsize=(10,5))
x = np.arange(1,11)
axes[0].plot(x,x**2,"g",lw=2)
axes[0].grid(True)
axes[0].set_title("Default Grid")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig,axes = plt.subplots(1,2,figsize=(10,5))
x = np.arange(1,11)
axes[0].plot(x,x**2,"g",lw=2)
axes[0].grid(True)
axes[0].set_title("Default Grid")
axes[1].plot(x,np.sin(x),"g",ls="--",lw=2)
axes[1].grid(True)
axes[1].set_title("Default Grid")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig,axes = plt.subplots(1,3,figsize=(10,5))
x = np.arange(1,11)
axes[0].plot(x,x**2,"g",lw=2)
axes[0].grid(True)
axes[0].set_title("Default Grid")
axes[1].plot(x,np.sin(x),"m",ls="--",lw=2)
axes[1].grid(True)
axes[1].set_title("Default Grid")
axes[2].plot(x,np.cos(x),"r",lw=2)
axes[2].grid(True)
axes[2].set_title("Default Grid")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig,axes=plt.subplots(1,3,figsize=(10,3))
x=np.arange(1,11)
axes[0].plot(x,x**2,"g",lw=2)
axes[0].grid(True,color="c")
axes[1].plot(x,x**2,"b",lw=2)
axes[1].grid(True,color="r")
axes[2].plot(x,x**2,"y",lw=2)
axes[2].grid(True,color="k")
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig,axes = plt.subplots(1,3,figsize=(10,5))
x = np.arange(1,11)
axes[0].plot(x,x**2,"g",lw=2)
axes[0].grid(True)
axes[0].set_title("Default Grid")
axes[1].plot(x,np.sin(x),"m",ls="--",lw=2)
axes[1].grid(True)
axes[1].set_title("Default Grid")
axes[2].plot(x,np.cos(x),"r",lw=2)
axes[2].grid(True,color='orange')
axes[2].set_title("Default Grid")
plt.show()
```
