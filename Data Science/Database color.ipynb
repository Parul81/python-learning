```python
# Coluring  in pandas:-
# Importing pandas as pd:-
# Know read the html_file
# Take the link from the google as:-
# "Https://en.wikipedia.org/wiki/List_of_highest-grossing_films"

import pandas as pd
df=pd.read_html("https://en.wikipedia.org/wiki/List_of_highest-grossing_films")
```

```python
# Take the database :-
# Call the database as "x":-
# "x=df[0]"
# Call "X"

x=df[0]
x
```

```python
# 1> Method:-
```

```python
# Make a def function as color_negative_red:-
# In the def function create a If-Else condition:-
# If the value is less than 2000 color it as red:-
# Elif the value is greater than 2000 color it as green:-
# Otherwise else color the rest as black:-
# And return the color as "%s":-

def color_negative_red(value):
    if value < 2000:
        color = 'red'
    elif value > 2000:
        color = 'green'
    else:
        color = 'black'
    return 'color: %s' % color
```

```python
# Styling the database "X" and applymap and set the color as red or green:-

x.style.applymap(color_negative_red,subset=["Year"])
```

```python
# Styling the properties on the behalf of back-ground color and color:-

x.style.set_properties(**{'background-color': 'pink','color': 'blue'})
```

```python
# Styling the database "X" as highlight_min and color it as purple:-

x.style.highlight_min(color="purple",axis=0)
```

```python
#  Styling the database "X" as highlight_max and color it as red:-

x.style.highlight_max(color="red",axis=0)
```

```python
# Setting background color as gradient color as cmap="Spectral":-

x.style.background_gradient(cmap="Spectral")
```

```python
# Making a def function as row_style :-
# Creating a If-Else condition:-
# If the year is less than "2015" than color the row as "pink":-
# Else color the rest year as "purple":-
# At the end style the database as "apply" at axix=1 or row:-

def row_style(row):
    if row.Year < 2015:
        return pd.Series('background-color: pink', row.index)
    else:
        return pd.Series('background-color: purple', row.index)

x.style.apply(row_style, axis=1)
```

```python
# Making a def function as column_style :-
# Creating a If-Else condition:-
# If the year is less than "2015" than color the column as "purple":-
# Else color the rest year as "orange":-
# At the end style the database as "apply" at axix=1 or column:-

def column_style(column):
    if column.Year < 2015:
        return pd.Series('background-color: purple', column.index)
    else:
        return pd.Series('background-color: orange', column.index)

x.style.apply(column_style, axis=1)
```
