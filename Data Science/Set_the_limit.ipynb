```python
import matplotlib.pyplot as plt
fig = plt.figure()
a1 = fig.add_axes([0,0,1,1])
import numpy as np
x = np.arange(1,10)
a1.plot(x,np.exp(x))
a1.set_title("Exp")
a1.set_ylim(0,100)
a1.set_xlim(0,10)
plt.show()
```
