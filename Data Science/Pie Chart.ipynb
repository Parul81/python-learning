```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis("Equal")
lang = ["Tableau","Python","ML","JAVA","DS"]
no_studs = [25,140,5,6,40]
ax.pie(no_studs,labels=lang)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis("Equal")
lang = ["Tableau","Python","ML","JAVA","DS"]
no_studs = [25,140,5,6,40]
ax.pie(no_studs,labels=lang,autopct='%1.2f%%')
plt.show()
```
