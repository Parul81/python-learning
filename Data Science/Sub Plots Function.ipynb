```python
import matplotlib.pyplot as plt
```

```python
fig,a = plt.subplots(2,2)
import numpy as np
```

```python
fig,a = plt.subplots(2,2)
import numpy as np
x = np.arange(1,5)
a[0][0].plot(x,x*x)
a[0][0].set_title("Square")
a[0][1].plot(x,np.sqrt(x))
a[0][1].set_title("Square Root")
a[1][0].plot(x,np.sin(x))
a[1][0].set_title("Sin")
a[1][1].plot(x,np.cos(x))
a[1][1].set_title("Cosine")
plt.show()
```
